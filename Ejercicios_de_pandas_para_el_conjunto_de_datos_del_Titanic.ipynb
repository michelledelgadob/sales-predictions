```python
#2.¿Cuál es el porcentaje de personas que sobrevivió? 
#(pista: encuentra la media de la columna de supervivencia)
media = df["Survived"].apply(lambda x: x)
denominador=media.mean()
```

```python
#3.¿Cuántos de cada sexo sobrevivieron?
filtro=df.loc[:,"Survived"]==1
df.loc[filtro, "Sex"].value_counts(dropna = False)
```

```python
#4.¿Cuál es el porcentaje de gente que sobrevivió que pagó un pasaje de menos de 10
filtro_2 = df.loc[:, "Fare"] < 10
len(df.loc[filtro & filtro_2,:])/len(df.loc[filtro,:])
```

```python
#5.¿Cuál es la edad promedio de quienes no sobrevivieron?
filtro_3=df.loc[:,"Survived"]==0
df.loc[filtro_3, "Age"].mean()
```
